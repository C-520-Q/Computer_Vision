## Preparing Our Data

```python
from torchvision import datasets
import torch
data_folder='../data'
# you want to download FMNIST to
fmnist=datasets.FashionMNIST(data_folder,download=True,train=True)
```

```python
tr_images=fmnist.data
tr_targets=fmnist.targets
```

```python
unique_values=tr_targets.unique()
print(f'tr_images & tr_targets:\n\tX-{tr_images.shape}'
      f'\n\tY-{tr_targets.shape}'
      f'\n\tY-Unique Values : {unique_values}')
print(f'TASK:\n\t{len(unique_values)} class Classification')
print(f'UNIQUE CLASSES:\n\t{fmnist.classes}')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
R,C=len(tr_targets.unique()),10
# 创建一个包含 R 行和 C 列的子图，设置图像大小为 10x10
fig, ax = plt.subplots(R, C, figsize=(10, 10))

# 遍历每个标签类和对应的绘图行
# label_class : 标签类索引
# plot_row : 每一行图对象
for label_class, plot_row in enumerate(ax):
    # 找到对应标签的所有图像索引
    label_x_rows = np.where(tr_targets == label_class)[0]
    
    # 遍历当前行中的每个绘图单元
    for plot_cell in plot_row:
        # 关闭网格并隐藏坐标轴
        plot_cell.grid(False)
        plot_cell.axis('off')
        
        # 从该标签类的图像索引中随机选择一个
        ix = np.random.choice(label_x_rows)
        x, y = tr_images[ix], tr_targets[ix]  # 获取随机选择的图像和对应标签
        
        # 在当前绘图单元中显示图像，使用灰度色彩映射
        plot_cell.imshow(x, cmap='gray')

# 自动调整子图参数以填充整个图像区域
plt.tight_layout()
```
